```python
#libraries
import pandas as pd
import scipy.stats as st
import numpy as np

```

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv")
df
```

- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

```python
 # We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.
# H0: The average HP of Dragon and Grass Pokemons is the same
# H1: The average HP of Dragon Pokemons is higher than Grass Pokemons
# We will use a one-tailed t-test to test this hypothesis

# First, we will filter the data to get the HP of Dragon and Grass Pokemons
df_hp = df[['Type 1', 'HP']]
df_hp = df_hp[(df_hp['Type 1'] == 'Dragon') | (df_hp['Type 1'] == 'Grass')]
df_hp
```

```python
# Now we will split the data into two groups
df_hp_dragon = df_hp[df_hp['Type 1'] == 'Dragon']
df_hp_grass = df_hp[df_hp['Type 1'] == 'Grass']
```

```python
# Now we will perform the t-test
t, p = st.ttest_ind(df_hp_dragon['HP'], df_hp_grass['HP'], equal_var=False)
print(f"t: {t}, p: {p}")
```

```python
# We will reject the null hypothesis if the p-value is less than 0.05
if p < 0.05:
    print("We reject the null hypothesis")
else:
    print("We fail to reject the null hypothesis")
```

```python
# The p-value is less than 0.05, so we reject the null hypothesis. This means that the average HP of Dragon Pokemons is higher than Grass Pokemons.
```

- We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.


```python
# We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.

# H0: The stats of Legendary and Non-Legendary Pokemons are the same
# H1: The stats of Legendary and Non-Legendary Pokemons are different
# We will use a two-tailed t-test to test this hypothesis

# First, we will filter the data to get the stats of Legendary and Non-Legendary Pokemons
df_stats = df[['Legendary', 'HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']]
df_stats
```

```python
# Now we will split the data into two groups
df_stats_legendary = df_stats[df_stats['Legendary'] == True]
df_stats_non_legendary = df_stats[df_stats['Legendary'] == False]

# Now we will perform the t-test for each stat
stats = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
for stat in stats:
    t, p = st.ttest_ind(df_stats_legendary[stat], df_stats_non_legendary[stat], equal_var=False)
    print(f"Stat: {stat}, t: {t}, p: {p}")
```

```python
# We will reject the null hypothesis if the p-value is less than 0.05
for stat in stats:
    if p < 0.05:
        print(f"Stat: {stat}, We reject the null hypothesis")
    else:
        print(f"Stat: {stat}, We fail to reject the null hypothesis")
```

```python
# The p-values are less than 0.05 for all stats, so we reject the null hypothesis. This means that the stats of Legendary and Non-Legendary Pokemons are different.
```

**We posit that houses close to either a school or a hospital are more expensive.**

- School coordinates (-118, 37)
- Hospital coordinates (-122, 34)

We consider a house (neighborhood) to be close to a school or hospital if the distance is lower than 0.50.

Hint:

```python
# **We posit that houses close to either a school or a hospital are more expensive.**

# School coordinates (-118, 37)
# Hospital coordinates (-122, 34)

# We consider a house (neighborhood) to be close to a school or hospital if the distance is lower than 0.50.
```

```python
# H0: The average price of houses close to a school or hospital is the same as the average price of houses not close to a school or hospital
# H1: The average price of houses close to a school or hospital is higher than the average price of houses not close to a school or hospital
# We will use a one-tailed t-test to test this hypothesis

# First, we will calculate the distance of each house to the school and hospital
df['distance_to_school'] = np.sqrt((df['longitude'] - (-118))**2 + (df['latitude'] - 37)**2)
df['distance_to_hospital'] = np.sqrt((df['longitude'] - (-122))**2 + (df['latitude'] - 34)**2)
```

```python
# Now we will filter the data to get the price of houses close to a school or hospital
df_price = df[['distance_to_school', 'distance_to_hospital', 'median_house_value']]
df_price = df_price[(df_price['distance_to_school'] < 0.50) | (df_price['distance_to_hospital'] < 0.50)]
df_price
```

```python
# Now we will split the data into two groups
df_price_close = df_price[(df_price['distance_to_school'] < 0.50) | (df_price['distance_to_hospital'] < 0.50)]
df_price_far = df_price[(df_price['distance_to_school'] >= 0.50) & (df_price['distance_to_hospital'] >= 0.50)]
```

```python
# Now we will perform the t-test
t, p = st.ttest_ind(df_price_close['median_house_value'], df_price_far['median_house_value'], equal_var=False)
print(f"t: {t}, p: {p}")
```

```python
# We will reject the null hypothesis if the p-value is less than 0.05
if p < 0.05:
    print("We reject the null hypothesis")
else:
    print("We fail to reject the null hypothesis")

```
